```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# These Libs are for stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
Response = pd.get_dummies(customer['response'], prefix='resp')
customer = pd.concat([customer, Response], axis=1)
customer
```

```python
channel = pd.get_dummies(customer['sales_channel'], prefix='sales')
customer = pd.concat([customer, channel], axis=1)
customer
```

```python
State = pd.get_dummies(customer['state'], prefix='stt')
customer = pd.concat([customer, State], axis=1)
customer
```

```python
Y = customer['total_claim_amount']
X = customer[['monthly_premium_auto', 'customer_lifetime_value','resp_No','resp_Yes','sales_Agent','sales_Branch','sales_Call Center','sales_Web']]
lm = linear_model.LinearRegression()
model = lm.fit(X,Y)
lm.score(X,Y) # This is the R^2
predictions = lm.predict(X)
print("R2 value is = ",round(r2_score(Y, predictions),2))
print("The intercept of the model is = ",lm.intercept_)
print("The coefficients of the model are = ",lm.coef_)
mse = mean_squared_error (Y, predictions)
mae = mean_absolute_error(Y, predictions)
print("The mse of the model is = ",mse)
print("The root mse of the model is = ",np.sqrt(mse))
print("The mean absolute error of the model is = ",mae)
```
